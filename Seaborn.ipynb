Rel plot
Redplot is same like scatter plot
here tips is dataframe

```python
sns.relplot(x="total_bill", y="tip", data=tips)
```

Differential in terms of Sex of total bill and tips- making many relation in one graph:hue function

To defentiate between the the time - style function

Seaborn helps in making showing relations in one graph 

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="sex", style="time", size='size')
```

To make the line plot inside rel plot we need to add kind function

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="sex", style="time", size='size', kind="line")
```

Example given by sir-
x=Total bill
y=tip
differentiate according to smoker and print 2 graphs according to time in lunch and dinner and differentiate according to smoker

col function helps in differentiating lunch and dinner and makes 2 graphs

```python
sns.relplot(x="total_bill", y="tip", data=tips,hue="smoker",col="time")
```

Categorical plot

```python
sns.catplot(x="day",y="total_bill",data=tips)
```

By default- Jetter is true, here we can see the points in sunday is not in one line they are scattered and by jitter we can bring them in one line

```python
sns.catplot(x="day",y="total_bill",data=tips,jitter=False)
```

Convert to Swarm plot by Kind function
Differentiate with the help of size

```python
sns.catplot(x="day",y="total_bill",data=tips,kind='swarm',hue='size')
```

To change the order

```python
sns.catplot(x='smoker',y='tip',data=tips,order=["No","Yes"])
```

Example:
x=total_bill,y=day
diff through sex, make violen plot  
make differentiation in one plot only, for every plot make only one violen plot

By default splot is False, it will give two plots and we want one plot only 

```python
sns.catplot(x="total_bill",y="day",data=tips,hue="sex",kind="violin")
```

As we wanted only one plot for one day, we are using Split as False so it will give us one plot

```python
sns.catplot(x="total_bill",y="day",data=tips,hue="sex",kind="violin", split='false')
```

here the inner line is percentile and if we want to remove that line we need to write inner=None

color=k will give all black color
We want to access the two different plots so

```python
sns.catplot(x='day', y='total_bill', kind='violin',data=tips,inner=None)
sns.swarmplot(x='day', y='total_bill',data=tips, color='k',size=3)
```
